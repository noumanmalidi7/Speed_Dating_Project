```python
# 6 main attributes
attributes = ['attr', 'sinc', 'intel', 'fun', 'amb', 'shar']
```

```python
#Average ratings match vs non-matches (female perspective)
female_matches = female_df[female_df['match'] == 1]
female_non_matches = female_df[female_df['match'] == 0]
```

```python
#Average ratings match vs non-matches (male perspective)
male_matches = male_df[male_df['match'] == 1]
male_non_matches = male_df[male_df['match'] == 0]
```

# CODE TO REVIEW

```python
print("\nAverage Attribute Ratings (Female Perspective):")
print("-" * 40)
attr_comparison_female = pd.DataFrame({
    'Matches': female_matches[attributes].mean(),
    'Non-Matches': female_non_matches[attributes].mean(),
    'Difference': female_matches[attributes].mean() - female_non_matches[attributes].mean()
})
print(attr_comparison_female.round(2).sort_values('Difference', ascending=False))
```

```python
print("\nAverage Attribute Ratings (Male Perspective):")
print("-" * 40)
attr_comparison_male = pd.DataFrame({
    'Matches': male_matches[attributes].mean(),
    'Non-Matches': male_non_matches[attributes].mean(),
    'Difference': male_matches[attributes].mean() - male_non_matches[attributes].mean()
})
print(attr_comparison_male.round(2).sort_values('Difference', ascending=False))
```

# PLOTLY GRAPH TO REVIEW

```python
# Interactive Plotly comparison
fig = go.Figure()

# Add female data
fig.add_trace(go.Bar(
    name='Female: Matches',
    x=attributes,
    y=attr_comparison_female['Matches'],
    marker_color='#4ECDC4',
    text=attr_comparison_female['Matches'].round(2),
    textposition='auto'
))

fig.add_trace(go.Bar(
    name='Female: Non-Matches',
    x=attributes,
    y=attr_comparison_female['Non-Matches'],
    marker_color='#FF6B6B',
    text=attr_comparison_female['Non-Matches'].round(2),
    textposition='auto'
))

fig.add_trace(go.Bar(
    name='Male: Matches',
    x=attributes,
    y=attr_comparison_male['Matches'],
    marker_color='#6A89CC',
    text=attr_comparison_male['Matches'].round(2),
    textposition='auto'
))

fig.add_trace(go.Bar(
    name='Male: Non-Matches',
    x=attributes,
    y=attr_comparison_male['Non-Matches'],
    marker_color='#FF9A8B',
    text=attr_comparison_male['Non-Matches'].round(2),
    textposition='auto'
))

fig.update_layout(
    barmode='group',
    title='Attribute Ratings: Gender Comparison',
    xaxis_title='Attributes',
    yaxis_title='Average Rating',
    height=500
)
fig.show()
```

# GENDER ANALYSIS

```python
print("\n👫 6. GENDER-SPECIFIC ANALYSIS: MEN VS WOMEN")
print("-" * 50)

# Calculate correlation with decisions for each gender
gender_comparison = []

for attr in attributes:
    # Female correlation with decision to say yes
    female_corr = female_df[attr].corr(female_df['dec'])
    
    # Male correlation with decision to say yes
    male_corr = male_df[attr].corr(male_df['dec'])
    
    # Difference
    diff = female_corr - male_corr
    
    gender_comparison.append({
        'Attribute': attr,
        'Women_Correlation': female_corr,
        'Men_Correlation': male_corr,
        'Difference': diff
    })

gender_df = pd.DataFrame(gender_comparison)

# Create parallel coordinates plot for gender comparison
fig = go.Figure(data=
    go.Parcoords(
        line=dict(
            color=gender_df.index,
            colorscale='Viridis',
            showscale=True,
            colorbar=dict(title="Attribute Index")
        ),
        dimensions=list([
            dict(range=[-1, 1],
                 label='Women', values=gender_df['Women_Correlation']),
            dict(range=[-1, 1],
                 label='Men', values=gender_df['Men_Correlation']),
            dict(range=[-0.5, 0.5],
                 label='Difference', values=gender_df['Difference'])
        ])
    )
)

fig.update_layout(
    title='<b>Gender Comparison: What Drives "Yes" Decisions?</b>',
    height=500
)

fig.show()

print("💡 Teaching: Parallel coordinates help compare multiple variables")
print("💡 Insight: Men and women value different attributes!")
```
